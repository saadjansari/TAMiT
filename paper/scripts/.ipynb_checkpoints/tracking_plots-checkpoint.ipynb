```python
from pathlib import Path
from matplotlib import pyplot as plt
import numpy as np
import scipy.io as spio
import random
import pandas as pd
import seaborn as sns
```

```python
mainpath = Path('/Users/saadjansari/Documents/Projects/ImageAnalysis/SingleCell/Results/Paper/fig5')
mainpath.resolve()
strain_names = ['WT','klp6D', 'alp14D', 'cls1-36']
foldnames = ['monopolar_wt', 'monopolar_klp6d', 'monopolar_alp14d', 'monopolar_cls1']

df = pd.DataFrame()
df['name'] = strain_names
df['path'] = [ mainpath / relpath for relpath in foldnames]

num_strains = df.shape[0]
```

```python
all_tau = []
all_times = []
all_lens = []
all_len_mean = []

# Loop over each strain
for idx in np.arange(num_strains):
    
    # Find all dydata.mat files
    path_mats = list( df.path[idx].glob('*/dydata.mat') )

    # Initialize lifetimes and lengths arrays
    tau = np.zeros(0)
    lens = np.zeros(0)
    times = np.zeros(0)
    
    # Loop over all dynamic feature files
    for pth in path_mats:
        # Load mat file
        mat_contents = spio.loadmat(pth, squeeze_me=True)
        
        # Concatenate data
        if type(mat_contents['tau']) == np.ndarray:
            tau = np.concatenate((tau,mat_contents['tau']), axis=0)
            lens = np.concatenate( (lens,mat_contents['lens']),axis=0)
            times = np.concatenate( (times,mat_contents['times']),axis=0)
        else:
            tau = np.concatenate((tau, np.array( [mat_contents['tau']])), axis=0)
            lens = np.concatenate( (lens, np.array( mat_contents['lens'])),axis=0)
            times = np.concatenate( (times, np.array( mat_contents['times'])),axis=0)
    all_tau.append(tau)
    all_lens.append(lens)
    all_times.append(times)
    all_len_mean.append( [np.mean(i) for i in lens] )

df['lifetime'] = all_tau
df['length'] = all_lens
df['time_vals'] = all_times
df['mean_length'] = all_len_mean
```

```python
# Plot lifetimes
fig, ax = plt.subplots(1, num_strains, figsize=(5*num_strains,4))
for idx in np.arange(num_strains):
    ax[idx].hist(df.lifetime[idx], bins=np.arange(0,50,2), color='#0504aa',
                                alpha=0.7, rwidth=0.85)
    ax[idx].set(xlabel='lifetime (sec)', ylabel='Count', xlim=[-0.1,30.1], title=df.name[idx])

# Plot lengths
fig, ax = plt.subplots(1, num_strains, figsize=(5*num_strains,4))
for idx in np.arange(num_strains):
    ax[idx].hist(df.mean_length[idx], bins=np.linspace(0,4,22), color='#0504aa',
                            alpha=0.7, rwidth=0.85)
    ax[idx].set(xlabel=r'Mean Length ($\mu m$)', ylabel='Count', title=df.name[idx])

#plt.savefig('{}/data_{}.pdf'.format(savepath, strain_name))
```

```python
df2 = pd.DataFrame()
name_list = []
lifetime_list = []
len_list = []
for idx in np.arange(num_strains):
    for idx2 in np.arange( len(df.lifetime[idx])):
        name_list.append( df.name[idx] )
        lifetime_list.append( df.lifetime[idx][idx2])
        len_list.append( df.mean_length[idx][idx2])
df2['strain'] = name_list
df2['lifetime'] = lifetime_list
df2['length'] = len_list

df3 = df2.dropna()
# ax = sns.violinplot(x="strain", y="lifetime", data=df3)
ax = sns.violinplot(x="strain", y="length", data=df3)
```
